```python
import numpy as np
```

```python
h = 1.0/50.0
tmax = 10.0
```

```python
numt = int(tmax/h) + 1
tvec = np.fromiter((i*h for i in range(numt)), np.double, count=numt)
```

```python
# yv = 0.5*(np.exp(tvec) + np.exp(-tvec)*(np.sin(2.0*tvec) - np.cos(2.0*tvec)))
yv = np.exp(-tvec)*np.sin(2.0*tvec)
ypv = -yv + 2.0*np.exp(-tvec)*np.cos(2.0*tvec)
```

```python
sol = np.fromiter((yv[i-1] + h*0.5*(ypv[i] + ypv[i-1]) for i in range(1, numt)), np.double, count=numt-1)
sol = np.insert(sol, 0, 0.0)
```

```python
# print(sol[0:5])
print(np.concatenate((sol[0:5], sol[-5:])))
print(np.concatenate((yv[0:5], yv[-5:])))
```

```python
print(np.log(np.abs(yv[1:] - sol[1:])/np.abs(yv[1:])))
```
