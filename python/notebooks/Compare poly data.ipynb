```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
import arma
```

```python
datadir = 'polycompdata'

Pmac = arma.matFromFile(datadir + '/bpoints.mac.arma')
Plin = arma.matFromFile(datadir + '/bpoints.lin.arma')

Bmac = arma.matFromFile(datadir + '/bvalues.mac.arma')
Blin = arma.matFromFile(datadir + '/bvalues.lin.arma')

Cmac = arma.matFromFile(datadir + '/coefficients.mac.arma')
Clin = arma.matFromFile(datadir + '/coefficients.lin.arma')
```

```python
rdiff = abs(Pmac[:,0].real - Plin[:,0].real)
print(sum(rdiff != 0))
idiff = abs(Pmac[:,0].imag - Plin[:,0].imag)
print(sum(rdiff != 0))

plt.subplot(1, 2, 1)
plt.plot(rdiff)
plt.subplot(1, 2, 2)
plt.plot(idiff)
plt.show()
```

```python
plt.plot(abs(Blin[:,0] - Bmac[:,0]))
plt.yscale('log')
```

```python
plt.plot(abs(Cmac[:,0].real - Clin[:,0].real), marker='o', linestyle='')
plt.yscale('log')
```
